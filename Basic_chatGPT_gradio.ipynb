```python
from transformers import pipeline

```

```python
generator=pipeline(task="text-generation",model="microsoft/DialoGPT-small")   #default model = openai-community/gpt2

def chatbot(user_input):
    response = generator(user_input, max_new_tokens=100,min_new_tokens=5, num_return_sequences=1,do_sample=True,temperature=0.6)
    return response[0]["generated_text"]
```

```python
chatbot("hello How are you")
```

```python
import gradio as gr

demo=gr.Interface(fn=chatbot,inputs="text",outputs="text",title="Basic ChatGpt2 bot")

demo.launch(share=True)
```
