```python
from transformers import pipeline

generator =pipeline(task='text-generation',model='distilgpt2')
```

```python
prompt = f"""
You are a data analyst.

Analyze the dataset below and give insights:

{data_text}

Insights:
"""
```

```python

text=generator(prompt,max_length=150, min_length=30, do_sample=False)
```

```python
def ask_data(question):
    prompt = f"""
    You are a data analyst.

    Dataset:
    {data_text}

    Answer clearly:

    Question: {question}
    Answer:
    """

    result = generator(prompt, max_length=120)
    return result[0]['generated_text']

print(ask_data("Which month has highest sales?"))
print(ask_data("Which region appears most?"))
```
