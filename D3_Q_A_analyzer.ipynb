# Question Answer context - H/F

```python
import transformers
```

```python
from transformers import pipeline
```

```python
ques_ans=pipeline("question-answering",model="distilbert/distilbert-base-cased-distilled-squad") #model ='deepset/roberta-base-squad2')
```

```python
predict=ques_ans(question="what is the capital of USA",
context="New York")

print(predict)
```
