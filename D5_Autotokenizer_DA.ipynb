```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
```

```python

import torch
!nvidia-smi  #Need to check T4 GPU is present to run mistralai it's a big model run in GPU
```

```python
from google.colab import userdata
HF_TOKEN=userdata.get('HF_TOKEN')
```

```python
#declare the model to be imported

model_name = "mistralai/Mistral-7B-Instruct-v0.1"

#import tokenizer

tokenizer = AutoTokenizer.from_pretrained(model_name,token=HF_TOKEN)

#import model
#load_in_4bit - loads and fit within memory short size the encoded value the mdoel , device _map=to GPU /CPU which ever is available hit the model downlaod

model=AutoModelForCausalLM.from_pretrained(model_name,
    device_map="auto",
    load_in_4bit=True,token=HF_TOKEN
)

# Create pipeline (EASIER + CLEANER)
pipe = pipeline(
    "text-generation", #taskname
    model=model,
    tokenizer=tokenizer,
    torch_dtype="auto"
)

prompt = f"""
You are a data analyst.

Here is the dataset summary:
{analysis}

Explain:
1. Key insights
2. Any anomalies
3. Business recommendations
"""

# Generate response
#prompt = create_prompt(df)

result = pipe(
    prompt,
    max_new_tokens=300,
    temperature=0.7,
    do_sample=True
)

print(result[0]["generated_text"])

```

### Steps to use Hugging Face Token:
1. **Generate a Hugging Face Token**: Go to your [Hugging Face profile settings](https://huggingface.co/settings/tokens) and create a new token with at least 'read' access.

3. The following code will then securely retrieve your token and use it when loading the model and tokenizer.
