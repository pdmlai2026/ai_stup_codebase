**AI model and AI task init**

```python
summarizer = pipeline("text-generation")
```

```python
text="The climate is nice and I want to "

summary=summarizer(text,max_new_tokens=80,min_new_tokens=20,do_sample=True,num_return_sequences=1,temperature=0.8)
```
