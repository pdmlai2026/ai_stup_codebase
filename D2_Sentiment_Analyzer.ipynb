# Program on Hugging Face distilbert model for Sentiment Analyzer task

```python
from transformers import pipeline
```

```python
senti_analyze= pipeline(task="sentiment-analysis",model ="distilbert/distilbert-base-uncased-finetuned-sst-2-english")
```

```python


ls_inpt=["I love databricks and ai ML","I don't love essay"]
print('Sentiment analysis is as follows :\n')
for ip_text in ls_inpt:
  sent_op=senti_analyze(ip_text)
  print(f'{ip_text} : denotes sentiment as - {sent_op[0]['label']}')

```
