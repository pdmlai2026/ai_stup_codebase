```python
import gradio as gr
```

```python
from transformers import pipeline

sent=pipeline(task="sentiment-analysis",model='distilbert/distilbert-base-uncased-finetuned-sst-2-english')

def sent_fn(text):
  if(text==""):
    return("Please enter some text")
  return(sent(text)[0]['label'])

```

```python
import gradio as gr


# Set a modern, clean theme
theme = gr.themes.Soft(
    primary_hue="blue",
    secondary_hue="slate",
)

with gr.Blocks(theme=theme, title="sentiment-analyse-basic-app") as demo:
    gr.Markdown("# Sentiment Analysis")
    gr.Markdown("Enter text below to check the sentiment : -")

    with gr.Row():
        # INPUT: Styled with label and placeholder
        input_box = gr.Textbox(
            label="Input Text",
            placeholder="Type your text here...",
            lines=3,
            elem_id="input-text"
        )

        # OUTPUT: Styled with label
        output_box = gr.Textbox(
            label="Result",
            placeholder="Output will appear here...",
            lines=3,
            interactive=False, # Makes output look distinct
            elem_id="output-text"
        )

    # BEAUTIFY: Add padding/margin using CSS if necessary (optional)
    # demo.css = ".gradio-container {padding: 20px !important;}"

    submit_btn = gr.Button("Analyze Sentiment", variant="primary")

    # Event handling
    submit_btn.click(fn=sent_fn, inputs=input_box, outputs=output_box)

#demo=gr.Interface(fn=sent_fn,inputs="text",outputs="text")

#demo.launch(share=True)

demo.launch(share=True,pwa=True)
```
