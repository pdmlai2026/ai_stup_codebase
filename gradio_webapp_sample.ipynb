```python
import gradio as gr

def greet(name):
    return f"Hi {name}, how are you!"

demo = gr.Interface(
    fn=greet,
    inputs="text",    # single-line text box
    outputs="text",   # display text
    title="Greet App",
    description="Type your name and get a friendly greeting."
)

# In Colab, set share=True to get a public link
demo.launch(share=True)
```
